# SETS
- 셋에 들어있는 데이터는 순서가 없습니다.
- 셋에는 동일한 값이 저장되지 않습니다. 우리가 배웠던 집합과 같습니다.
- 셋은 중괄호 { } 로 정의합니다.

```python
data = set()
data
```

```python
set_data = {1,3,4,1,2,3,4,12,5,2,4,1}
set_data
```

```python
# 리스트 -> 집합으로 캐스팅(형변환)
list_data = [1,3,4,1,2,3,4,12,5,2,4,1]
list_data
```

```python
set(list_data)
```

```python
# 튜플 -> 집합으로 캐스팅(형변환)
tuple_data = (1,3,4,1,2,3,4,12,5,2,4,1)
tuple_data
```

```python
set(tuple_data)
```

```python
set_data = set(tuple_data)
```
