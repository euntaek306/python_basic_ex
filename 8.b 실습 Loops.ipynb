# <font color='blue'> 음수 입력하거나 9단보다 큰 단을 입력하면, “잘못입력하였습니다.” 하고, 다시 단을 입력 받습니다.
# <font color='blue'> 단, 0 을 입력받으면, 프로그램을 종료하세요.

# <font color='blue'>          ...


```python
# 2를 입력받으면, 2단에서 9단까지 출력하세요.
# 8을 입력받으면, 8단에서 9단까지 출력하세요.
# 사용자로부터 5를 입력 받음.
num = int('5')

# 5 x 1 = 5
# 5 x 2 = 10
# 5 x 3 = 15
# ...
# 5 x 9 = 45
# ...
# 9 x 1 = 9
# ...
# 9 x 9 = 81
```

```python
num = int('5')
for i in range(num, 9+1): #5~9단까지
  for j in range(1, 9+1):
    print(f"{i} x {j} = {i * j}")
  print('-'*10)
```

```python
# 반복문 구현
# 음수 입력하거나 9단보다 큰 단을 입력하면, “잘못입력하였습니다.” 하고, 다시 단을 입력 받습니다.
# 단, 0 을 입력받으면, 프로그램을 종료하세요.

while True:
  input_val = input("0~9 숫자를 입력하세요.")
  if input_val.isalpha() or int(input_val) < 0 or int(input_val) > 9:
    pass
  else:
    num = int(input_val)
    # 0일 경우 -> 종료
    if num == 0:
      print("종료")
      break
    else:
      # 0이 아닌 경우 -> 곱셈 출력
      for i in range(num, 9+1): #5~9단까지
        for j in range(1, 9+1):
          print(f"{i} x {j} = {i * j}")
        print('-'*10)
```

# <font color='blue'> 12. 해당 sms 문장에 스팸으로 등록한 단어가 있는지 검사하세요.
## 스팸 단어 리스트 = [ '대출', '금리', '도박', '잭팟']
## 검사한 후, 스팸 단어가 있으면, "스팸"을 출력하고, 없으면 "스팸 아님"을 출력하세요.

## 문장 예 ) "최저 금리로 당일 송금, 대출 100만~3000만원 가능"

```python
spam_words = [ '대출', '금리', '도박', '잭팟']

# sms = "최저 금리로 당일 송금, 대출 100만~3000만원 가능"
sms = "동생에게 100만원 입금 완료"
```

```python
# 스팸 메세지인지 검사하기,
# 스팸 단어가 있으면,
# "스팸"을 출력하고,
# 없으면
# "스팸 아님"을 출력하세요.
for word in spam_words:
  if word in sms:
    print("스팸 메세지 입니다")
    break
  else:
    pass
print("스팸 메세지가 아닙니다")


```

```python
spam_words = [ '대출', '금리', '도박', '잭팟']
# sms = "최저 금리로 당일 송금, 대출 100만~3000만원 가능"
sms = "동생에게 100만원 입금 완료"
spam_cnt = 0
for word in spam_words:
  if word in sms:
    spam_cnt = spam_cnt + 1
    break
  else:
    pass
if spam_cnt > 0:
  print("스팸 메세지가 아닙니다")
else:
  print("스팸 메세지가 아닙니다")
```
