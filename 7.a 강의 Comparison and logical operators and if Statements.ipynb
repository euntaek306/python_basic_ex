```python
my_score = 50
cut_line = 60
# 참인 경우만 실행할 때
if my_score > cut_line :
  print("합격")
```

```python
# 참인 경우와 거짓인 경우 각각에 판단해서 실행
if my_score > cut_line :
  print("합격")
else:
  print("불합격")
```

```python
score = 70
score >= 60 and score < 80
```

```python
# 60 미만 : 더 공부하세요.
# 60~80미만 : 합격 했어요.
# 80~이상 : 최고 예요.
score = 50
cut_line = 60
if score < cut_line :
  print("더 공부하세요.")
  # 또 다른 조건 score >= 60 and score < 80건
elif score >= 60 and score < 80 :
  print("합격 했어요")
else:
  print("최고 예요.")
```
